 ## Reading from CSV/TSV Files
 
 > <b>CSV:</b> A text file in which the values are separated by a comma or a tab character is called a CSV or a TSV file. Each line of the file is a data record and each record consists of one or more fields, separated by a specific character called separator. A CSV/TSV file is typically used to store tabular data (numbers and text), in which each line will have the same number of fields.
 
The pd.read_csv() method is used to read a comma-separated file into a DataFrame.

<strong>pd.read_csv(fname, delimiter=None, header='infer', skiprows=None , nrows=None , usecols=None,  footer='',...)</strong>

```python
# Read just 10 rows from the csv file by passing the number of rows to read to `nrows` argument
df = pd.read_csv('data.csv', nrows=10)
print(df.shape)
df.head()
```

```python
# Read specific columns from the csv file by passing a list of column names to `usecols` argument
df = pd.read_csv('data.csv', usecols= ['StudentID', 'Group','English'])
df.shape
df.head()
```

## I/O with EXCEL Files

>XLSX: XLSX is a Microsoft Excel Open XML file format. It also comes under the Spreadsheet file format. It is an XML-based file format created by Microsoft Excel. In XLSX data is organized under the cells and columns in a sheet. Each XLSX file may contain one or more sheets. So a workbook can contain multiple sheets
